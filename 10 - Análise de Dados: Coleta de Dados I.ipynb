## 1\. Excel para CSV

Utilizando o pacote Python `openpyxl` visto em aula, extraia os seguintes as colunas `id`, `sexo` e `idade` para dos clientes inadimplentes (`default = 1`) e solteiros (`estado_civil = 'solteiro'`). Salves os dados extraídos no arquivo csv `credito.csv` separado por `;`. Exemplo do cabeçalho e das três primeiras linhas:

```

767712558;59;M
```

**Dica:** O arquivo csv `credito.csv` deve ter 669 linhas, contando com o cabeçalho.

````python
from openpyxl import load_workbook
import csv

arquivo = load_workbook(filename='credito.xlsx')
planilha = arquivo.active



cabecalho = next(planilha.values)
indice_inad = cabecalho.index('default')
indice_estcivil = cabecalho.index('estado_civil')


with open(file='./arquivo.csv', mode='w', encoding='utf8') as arquivo:
  colunas = ['id', 'sexo', 'idade']
  escritor_csv = csv.writer(arquivo, delimiter=';')
  escritor_csv.writerow(colunas)
  for linha in planilha.values:
    if linha[indice_inad] == 1 and linha[indice_estcivil] == 'solteiro':
        dados = []
        for coluna in colunas:
          dados.append(linha[cabecalho.index(coluna)])
        escritor_csv.writerow(dados)





````

**Dica:** Sua solução deve gerar o dicionário Python `credito` igual ao exemplo mas a ordem dos elementos pode variar tranquilamente.

**Dica:** Uma excelente forma de remover elementos duplicados de uma lista é convertê-la para `set` e depois para `list` novamente.



````python
from openpyxl import load_workbook
import json

arquivo = load_workbook(filename='credito.xlsx')
planilha = arquivo.active


def extrair_coluna(planilha, coluna):
    cabecalho = next(planilha.values)
    indice = cabecalho.index(coluna)
    return [linha[indice] for linha in planilha.values if linha[indice] != coluna]

colunas = ['escolaridade', 'tipo_cartao']
credito = {}

for i in range(0, len(colunas)):
    credito.update({colunas[i]:list(set(extrair_coluna(planilha, colunas[i])))})

credito_json = json.dumps(credito, indent=4)
print(credito_json)
````

## 3\. BÔNUS: Texto para CSV

No arquivo de texto `ebac.txt` você encontra o texto presente no rodapé da página de cursos da EBAC ([link](https://ebaconline.com.br/)).



**Arquivo TXT:** ebac.txt

````python
%%writefile ebac.txt
MÍDIAS SOCIAIS
Instagram, Facebook, Youtube, LinkedIn 

CURSOS
Software, Design, Marketing, Audiovisual, Programação & Data, Games

WEBINARS
Próximos, Anteriores

SOBRE
Sobre nós, Centro de carreiras, Vagas

CONTATO
WhatsApp +55 (11) 4200-2991
Telefone +55 (11) 3030-3200

BLOG
Design, Audiovisual, Marketing
````

Extraia os números de contato do arquivo texto `ebac.txt` e salve-os no arquivo csv `contato_ebac.csv` com o separador `;` no seguinte formato:

```
tipo;numero
whatsapp;+551142002991
telefone;+551130303200
```

**Nota:** Escreva o código da sua solução abaixo em uma ou mais células, você não precisa enviar o arquivo csv gerado.

````python
import re
with open(file='./ebac.txt', mode='r', encoding='utf8') as arquivo:
  linhas = arquivo.readlines()

linhas = filter(lambda linha: linha != '\n', linhas)
linhas = map(lambda linha: linha.strip(), linhas)
linhas = list(linhas)

#Formata uma string para remover quaisquer caracteres que não sejam numéricos ou o caractere "+"
def filtrar_numero(string):
  s = re.sub(r"[^+Z0-9]", "", string)
  return s

#Filtra uma string e retorna verdadeiro caso encontre a palavra "whatsapp" nessa string em diferentes formas de escrita
def filtrar_whatsapp(string):
  filtro_whatsapp = ["WhatsApp", "whatsapp", "Whatsapp"]
  for i in range(0, len(filtro_whatsapp)):
    if filtro_whatsapp[i] in string:
      return True
    
#Filtra uma string e retorna verdadeiro caso encontre a palavra "telefone" nessa string em diferentes formas de escrita
def filtrar_telefone(string):
  filtro_telefone = ["Telefone", "telefone"]
  for i in range(0, len(filtro_telefone)):
    if filtro_telefone[i] in string:
      return True    

whatsapp = list(filter(filtrar_whatsapp, linhas))
telefone = list(filter(filtrar_telefone, linhas))



with open(file='./ebac.csv', mode='w', encoding='utf8') as arquivo:
  escritor_csv = csv.writer(arquivo, delimiter=';')
  escritor_csv.writerow(['tipo','numero'])
  for i in range(0, len(whatsapp)):
    escritor_csv.writerow(['whatsapp', filtrar_numero(whatsapp[i])])
  for i in range(0, len(telefone)):
    escritor_csv.writerow(['telefone', filtrar_numero(telefone[i])])

````
